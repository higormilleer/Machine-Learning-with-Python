# K-Means Clustering Example

Let's make some fake data that includes people clustered by income and age, randomly:

```python
from numpy import random, array

#Create fake income/age clusters for N people in k clusters
def createClusteredData(N, k): #criar grupos de n pessoas
    random.seed(10)# uma semente aletorio de 10 em k grupos
    pointsPerCluster = float(N)/k
    X = []#lista inicialmente vazia
    for i in range (k      #para cada grupos para i no intervalo de  
        incomeCentroid = random.uniform(20000.0, 200000.0) #criar um entroide aleatorio de renda entre os dois valores colocados
        ageCentroid = random.uniform(20.0, 70.0) #centroide de idade aleatoio de 20 a 70 anos uniformemente 
        for j in range(int(pointsPerCluster #crianco um grafico de dispersao que mostrara renda x idade para n pessoas em k grupos 
            X.append([random.normal(incomeCentroid, 10000.0), random.normal(ageCentroid, 2.0)]) #com desvio padrao de 10 mil de renda e 2 de idade
    X = array(X)
    return X
```

We'll use k-means to rediscover these clusters in unsupervised learning:

```python
%matplotlib inline

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from numpy import random

data = createClusteredData(100, 5) #100 pessoas em 5 grupos direntes

model = KMeans(n_clusters=5)

# Note I'm scaling the data to normalize it! Important for good results.
model = model.fit(scale(data))

# We can look at the clusters each data point was assigned to
print(model.labels_)

# And we'll visualize it:
plt.figure(figsize=(8, 6)) #criando o grid
plt.scatter(x=data[:,0], y=data[:,1], c=model.labels_.astype(float)) #atribui cor aos rotulos anteriormente criados
plt.show()
```
