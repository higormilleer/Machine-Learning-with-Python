# Linear Regression

Let's fabricate some data that shows a roughly linear relationship between page speed and amount purchased:

```python
%matplotlib inline
import numpy as np
from pylab import *

pageSpeeds = np.random.normal(3.0, 1.0, 1000)  # ele gera 1000 valores aleatórios, para representar a velocidade da pagina
purchaseAmount = 100 - (pageSpeeds + np.random.normal(0, 0.1, 1000)) * 3  # faz o calculo do valor de compras com base na velocidade da página, adicionando um ruido aleatorio pequeno, simulando uma relacao inversa.

scatter(pageSpeeds, purchaseAmount)  # Cria um gráfico de dispersão (scatter plot) para visualizar a relação entre velocidades de página e valores de compra.
```

As we only have two features, we can keep it simple and just use scipy.state.linregress:

```python
from scipy import stats #fornece recursos estatisticos

slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchaseAmount)  # Realiza uma regressao linear nos dados, retornando:
# slope: a inclinacao da linha de regressao.
# intercept: o intercepto da linha no eixo y.
# r_value: o coeficiente de correlacao.
# p_value: o valor-p para testar a hipotese nula.
# std_err: o erro padrap da estimativa da inclinacao.
```

Not surprisngly, our R-squared value shows a really good fit:

```python
r_value ** 2 #calcul o coeficiente de determinação, o Rˆ2, medindo a variacao dos dados (sendo de 0 a 1) 
```

Let's use the slope and intercept we got from the regression to plot predicted values vs. observed:

```python
import matplotlib.pyplot as plt

def predict(x):  #define uma funcao para prever valores de y com base na regressão linear
    return slope * x + intercept  #aplica a equacao da linha de regressao
fitLine = predict(pageSpeeds)  # gera os valores previstos para a linha de regressao 

plt.scatter(pageSpeeds, purchaseAmount)  #cria um grefico de dispersao mostrando a relacao entre pageSpeeds e purchaseAmount.
plt.plot(pageSpeeds, fitLine, c='r')  #faz a plotagem da linha de regressão linear em vermelho sobre os pontos do scatter plot.
plt.show()  # Mostra o grafico
```

## Activity

Try increasing the random variation in the test data, and see what effect it has on the r-squared error value.

```python
# Aumenta o ruido aleatorio em purchaseAmount
pageSpeeds = np.random.normal(3.0, 1.0, 1000)  #mantem a mesma distribuicao para pageSpeeds
purchaseAmount = 100 - (pageSpeeds + np.random.normal(0, 1.0, 1000)) * 3  #aumenta o ruido 

#realiza a regressao linear
slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchaseAmount)

# Imprime o valor atualizado de R-quadrado, tendo a dispersao menor que 0,5
print("R-quadrado:", r_value ** 2)

# Plota com a variacao aumentada
def predict(x):  #define a funcao de previsao baseada na linha de regressao
    return slope * x + intercept

fitLine = predict(pageSpeeds)  #calcula a linha de ajuste usando os valores de pageSpeeds

plt.scatter(pageSpeeds, purchaseAmount)  #plota o grafico de dispersao dos dados
plt.plot(pageSpeeds, fitLine, c='r')  #plota a linha de regressao 
plt.show()  
```
